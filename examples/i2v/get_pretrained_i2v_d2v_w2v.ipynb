```python
from EduNLP import get_pretrained_i2v
import numpy as np
```

# 通过i2v从模型库中下载w2v模型

```python
i2v = get_pretrained_i2v("w2v_lit_256", "../../data/w2v")
item_vector, token_vector = i2v(["有学者认为：‘向西方学习’，必须适应和结合实际才有作用"])

print("item_vector : \n", np.array(item_vector).shape, item_vector)
print("token_vector : \n", np.array(token_vector).shape, token_vector)
```

# get_pretrained_i2v

## 概述

使用 EduNLP 项目组给定的预训练模型将给定的题目文本转成向量。

- 优点：简单方便。
- 缺点：只能使用项目中给定的模型，局限性较大。


## 导入功能块

```python
from EduNLP import get_pretrained_i2v
```

## 输入

类型：str  
内容：题目文本 （text）

```python
item = [{
"如图来自古希腊数学家希波克拉底所研究的几何图形．此图由三个半圆构成，三个半圆的直径分别为直角三角形$ABC$的斜边$BC$, 直角边$AB$, $AC$.$\bigtriangleup ABC$的三边所围成的区域记为$I$,黑色部分记为$II$, 其余部分记为$III$.在整个图形中随机取一点，此点取自$I,II,III$的概率分别记为$p_1,p_2,p_3$,则$\SIFChoice$$\FigureID{1}$"
}]
```

## 模型选择与使用

根据题目所属学科选择预训练模型：    

 预训练模型名称 | 模型训练数据的所属学科   
 -------------- | ----------------------   
 d2v_all_256    | 全学科                   
 d2v_sci_256    | 理科                     
 d2v_eng_256    | 英语                     
 d2v_lit_256    | 文科                      



```python
save_dir =  "./pubulic_model"

# w2v
i2v = get_pretrained_i2v("w2v_all_256", mdoel_dir=save_dir)

# d2v
i2v = get_pretrained_i2v("d2v_all_256", mdoel_dir=save_dir)

# emlo
i2v = get_pretrained_i2v("elmo_test", mdoel_dir=save_dir)

# bert
i2v = get_pretrained_i2v("luna_bert", mdoel_dir=save_dir)

# disenq
i2v = get_pretrained_i2v("disenq_pub_128", mdoel_dir=save_dir)
```

- 注意：
    默认的 EduNLP 项目存储地址为根目录(`~/.EduNLP`)，模型存储地址为项目存储地址下的 `model` 文件夹。您可以通过修改下面的环境变量来修改模型存储地址：
    - EduNLP 项目存储地址：`EDUNLPPATH = xx/xx/xx`
    - 模型存储地址：`EDUNLPMODELPATH = xx/xx/xx`

```python
for item in item:
  item_vector, token_vector = i2v(item)
	# or
  # item_vector = i2v.infer_item_vector(item)
	# token_vector = i2v.infer_item_vector(item)
    
```
